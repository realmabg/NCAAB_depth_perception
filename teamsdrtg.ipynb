```python
import pandas as pd
```

```python
stats2021 = pd.read_csv("BartTorvik Team Stats/2021_team_results.csv")
stats2022 = pd.read_csv("BartTorvik Team Stats/2022_team_results.csv")
stats2023 = pd.read_csv("BartTorvik Team Stats/2023_team_results.csv")
stats2024 = pd.read_csv("BartTorvik Team Stats/2024_team_results.csv")
stats2025 = pd.read_csv("BartTorvik Team Stats/2025_team_results.csv")
```

```python
stats2025["Year"] = 2025
stats2024["Year"] = 2024
stats2023["Year"] = 2023
stats2022["Year"] = 2022
stats2021["Year"] = 2021
```

```python
stats_dfs = {
    2025: stats2025,
    2024: stats2024,
    2023: stats2023,
    2022: stats2022,
    2021: stats2021
}

for year, df in stats_dfs.items():
    new_columns = list(df.columns[1:]) + ['Fun stat']
    df.columns = new_columns

    # Set index name
    df.index.name = "rank"

    # Add year column
    df["Year"] = year
```
