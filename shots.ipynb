```python
import pandas as pd
```

```python
shots_df = pd.read_csv("pbp_shots.csv")
shots_df
```

```python
comp_to_season = {
    27693: 2021,
    30629: 2022,
    33533: 2023,
    36046: 2024,
    38409: 2025
}
shots_df["season"] = shots_df["competitionId"].map(comp_to_season)
```

```python
shots_df
```

```python
shots_df["homeMarket"].unique()
```

```python
jump_shots_list = ['jumpshot', 'stepbackjumpshot', 'hookshot',
       'floatingjumpshot', 'pullupjumpshot',
       'turnaroundjumpshot']
```

```python
jump_shots_df = shots_df[shots_df["subType"].isin(jump_shots_list)]
jump_shots_df
```

```python
ucsd_home_game_jump_shots = jump_shots_df[jump_shots_df["homeMarket"] == "UC San Diego"]
ucsd_home_game_jump_shots
```

```python
ucsd_home_game_jump_shots_left = ucsd_home_game_jump_shots[(ucsd_home_game_jump_shots["side"] == "LEFT") & (ucsd_home_game_jump_shots["teamId"]!=104355)]["success"].mean()
ucsd_home_game_jump_shots_left
```

```python
ucsd_home_game_jump_shots_right = ucsd_home_game_jump_shots[(ucsd_home_game_jump_shots["side"] == "RIGHT") & (ucsd_home_game_jump_shots["teamId"]!=104355)]["success"].mean()
ucsd_home_game_jump_shots_right
```
